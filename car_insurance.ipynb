# **Create a new column with the duration of each call in seconds**

```python
# Convert 'CallStart' and 'CallEnd' columns to datetime objects
data1['CallStart'] = pd.to_datetime(data1['CallStart'])
data1['CallEnd'] = pd.to_datetime(data1['CallEnd'])


data1['call_duration']= data1['CallEnd']- data1['CallStart']
```
